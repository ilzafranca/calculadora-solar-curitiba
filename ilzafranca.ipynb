```python
def calcular_sistema_solar(consumo_mensal_kwh, tipo_imovel):
    # Dados base para Curitiba
    radiacao_media = 4.3  # kWh/m¬≤/dia
    dias_mes = 30
    eficiencia_sistema = 0.75  # perdas do sistema (~25%)

    # Estimativa de gera√ß√£o por placa (placa de 550W)
    potencia_placa_kw = 0.55
    geracao_mensal_por_placa = potencia_placa_kw * radiacao_media * dias_mes * eficiencia_sistema

    # N√∫mero de placas necess√°rio
    num_placas = consumo_mensal_kwh / geracao_mensal_por_placa
    num_placas = int(num_placas) + 1  # arredondando pra cima

    # Custo estimado por tipo de im√≥vel
    if tipo_imovel.lower() == 'resid√™ncia':
        custo_placa = 1600  # R$ por placa instalada
    elif tipo_imovel.lower() == 'com√©rcio':
        custo_placa = 1500
    elif tipo_imovel.lower() == 'pr√©dio':
        custo_placa = 1450
    else:
        return "Tipo de im√≥vel inv√°lido."

    investimento_total = num_placas * custo_placa

    # Economia mensal estimada (100% de compensa√ß√£o)
    economia_mensal = consumo_mensal_kwh * 0.9  # supondo tarifa de R$0,90/kWh
    retorno_em_meses = investimento_total / economia_mensal
    retorno_em_anos = retorno_em_meses / 12

    # Resultados
    print(f"‚ö°Ô∏è Consumo mensal: {consumo_mensal_kwh} kWh")
    print(f"üè† Tipo de im√≥vel: {tipo_imovel}")
    print(f"üîã Placas necess√°rias: {num_placas}")
    print(f"üí∞ Investimento estimado: R$ {investimento_total:,.2f}")
    print(f"üìâ Economia mensal: R$ {economia_mensal:,.2f}")
    print(f"‚è≥ Retorno estimado: {retorno_em_anos:.1f} anos")

# Exemplo de uso:
calcular_sistema_solar(consumo_mensal_kwh=450, tipo_imovel='resid√™ncia')
```
